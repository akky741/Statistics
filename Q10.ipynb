```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(r"Q7.csv",index_col=0)
```

```python
data
```

```python
f, ax = plt.subplots(figsize=(15, 5))


plt.subplot(1, 3, 1)
plt.boxplot(data.Points)
plt.title('Points')


plt.subplot(1, 3, 2)
plt.boxplot(data.Score)
plt.title('Score')


plt.subplot(1, 3, 3)
plt.boxplot(data.Weigh)
plt.title('Weigh')

plt.show()
```

```python
f, ax = plt.subplots(figsize=(15, 5))


plt.subplot(1, 3, 1)
plt.hist(data.Points,edgecolor = 'black')
plt.title('Points')


plt.subplot(1, 3, 2)
plt.hist(data.Score,edgecolor = 'black')
plt.title('Score')


plt.subplot(1, 3, 3)
plt.hist(data.Weigh,edgecolor = 'black')
plt.title('Weigh')

plt.show()
```
