```python
dataa = pd.read_csv("Cars.csv")
dataa
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean, cov = [0, 0], [(1, .6), (.6, 1)]
x, y = np.random.multivariate_normal(mean, cov, 100).T
y += x + 1

f, ax = plt.subplots(figsize=(6,6))

ax.scatter(x,y, c="red")
ax.set(xlim=(-3,3),ylim=(-3,3))

#draw diagonal
diag_line, = ax.plot(ax.get_xlim(), ax.get_ylim(), ls = "--", c = "blue")

def on_change(axes):
    x_lims = ax.get_xlim()
    y_lims = ax.get_ylim()
    diag_line.set_data(x_lims,y_lims) 


#connects callbacks to your axes

ax.callbacks.connect("xlim_changed", on_change)
ax.callbacks.connect("ylim_changed", on_change)

plt.show()


```

```python
import statsmodels.api as sm
sm.qqplot(dataa['MPG'])
plt.xlabel('MPG', color ='red')
plt.savefig('MPG of cars.png')
plt.show()
```

```python
import scipy.stats as stats
stats.probplot(dataa['MPG'], dist="norm", plot=plt)
plt.xlabel('MPG', color ='red')
plt.savefig('MPG of cars.png')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.distplot(dataa['MPG'],kde=True, bins =10)
plt.show()
```

MPG of cars follows normal distribution
